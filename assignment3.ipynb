3. Python program to insert a number to any position in a list (Get the number
to be inserted and position from the user)

```python
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

number_to_insert = int(input("Enter the number to insert: "))

position = int(input("Enter the position (index) to insert the number: "))

my_list.insert(position, number_to_insert)

print(f"Modified list after inserting {number_to_insert} at position {position}: {my_list}")
```

4. Write a program to rename a key city to a location in the following dictionary.
sample_dict = { "name": "Kelly", "age":25, "salary": 8000, "city": "New york"}
