1. Grocery Billing
A shopkeeper wants to calculate the total bill for two items. Write a program that takes
the price and quantity of each item and prints the total bill amount.
Sample Input:
Item1 price: 45
Item1 quantity: 3
Item2 price: 30
Item2 quantity: 2
Sample Output:
Total Bill: 195

```python
# Get input for Item 1
price_item1 = float(input("Enter the price of item 1: "))
quantity_item1 = int(input("Enter the quantity of item 1: "))

# Get input for Item 2
price_item2 = float(input("Enter the price of item 2: "))
quantity_item2 = int(input("Enter the quantity of item 2: "))

# Calculate total cost for each item
total_cost_item1 = price_item1 * quantity_item1
total_cost_item2 = price_item2 * quantity_item2

# Calculate the total bill
total_bill = total_cost_item1 + total_cost_item2

# Print the total bill amount
print(f"\nTotal bill amount: ${total_bill:.2f}")
```

Write a program to calculate simple interest using the formula SI = (P Ã— R Ã— T) / 100.

```python
principal = float(input("Enter the principal amount (P): "))
rate = float(input("Enter the annual interest rate (R in %): "))
time = float(input("Enter the time period (T in years): "))

simple_interest = (principal * rate * time) / 100

print(f"Principal Amount: {principal:.2f}")
print(f"Interest Rate: {rate:.2f}%")
print(f"Time Period: {time:.2f} years")
print(f"Simple Interest: {simple_interest:.2f}")
```
