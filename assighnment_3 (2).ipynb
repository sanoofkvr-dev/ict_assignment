# Q1

```python
def max_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
num1 = 32
num2 = 45
num3 = 95
print('the maximum number is',max_of_three(num1,num2,num3))
```

# Q2

```python
def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total
sample_list = [19, 14, 20, 20, 6]
result = sum_list(sample_list)
print(result)     


```
